```python
from collections import Counter
```

# ***Which ingredient is the most used***

```python
df_type['cleaned_ingredients'] = df_type['ingredients'].apply(lambda x: " ".join(x.lower().strip().replace(" ", "").split()))
cnt=Counter()
df_type.cleaned_ingredients.str.split(',|;').apply(cnt.update)
d=dict(cnt)
```

```python
print(" Ingredientul cel mai utilizat ")

max(d.items(), key=lambda k: k[1])
```

```python
from tabulate import tabulate
table_data = [(key, value) for key, value in d.items()]

# Afișați tabelul folosind tabulate
print(tabulate(table_data, headers=['Ingredient', 'Count'], tablefmt='psql'))
```

# ***How many types of cheese does the restaurant uses***

```python
import re

cheese_types = []
s=0
for row in df_type['ingredients']:
    ingredients = re.split(r',|;', row)
    for ingredient in ingredients:
        if 'cheese' in ingredient.lower():
            cheese_types.append(ingredient.strip())

unique_cheese_types = list(set(cheese_types))
for cheese_type in unique_cheese_types:
    s+=1
df = pd.DataFrame({'Cheese Type': unique_cheese_types})
print(tabulate(df, headers='keys', tablefmt='psql'))

```

```python
print(f'Restaurant use {s} types of cheese')
```

## ***How much pizza has the restaurant sold for each type and size***

```python
df_merged = pd.merge(df_orderdet, df_pizza, on="pizza_id")
df_merged = pd.merge(df_merged, df_type, on="pizza_type_id")

df_grouped = df_merged.groupby("name")["quantity"].sum().reset_index()

print(tabulate(df_grouped, headers='keys', tablefmt='psql'))
```

# ***Top 5 most ordered pizza***

```python
top_5_pizzas = df_grouped.sort_values(by="quantity", ascending=False).head(5)

print(tabulate(top_5_pizzas, headers='keys', tablefmt='psql'))
```

# ***Total revenue generated by the restaurant in 2015***

```python
df_merged['revenue']= df_merged['price'] * df_merged['quantity']
df_merged['revenue'].sum()
```

```python
df_merged = pd.merge(df_merged, df_orders, on="order_id")
df_merged.info()
```

# ***Revenue generated each month of 2015***

```python
df_merged['date'] = pd.to_datetime(df_merged['date'])

df_merged['month'] = df_merged['date'].dt.month

revenue_per_month = df_merged.groupby(df_merged['month'])['revenue'].sum()
```

```python
plt.figure(figsize=(10, 6))
revenue_per_month.plot(kind='bar', color='skyblue')
plt.title('Revenue per month')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

It can be seen that the income is constant throughout the year, there are no big differences from one month to another.

The most orders were made in July, exceeding 70000, and the fewest in September and October, just under 65000

# ***How many orders does the restaurant handles in a day on average?***

```python
orders_per_day = df_merged.groupby(df_merged['date'].dt.date)['order_id'].nunique()
print(orders_per_day.mean())
```

# ***Is there a day of the week that is more crowded?***

```python
orders_per_day
```

```python
df_merged['day_of_week'] = df_merged['date'].dt.day_name()
```

```python
average_orders_per_day_of_week = df_merged.groupby(['day_of_week', 'date'])['order_id'].nunique().groupby('day_of_week').mean()
days_of_week_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
average_orders_per_day_of_week = average_orders_per_day_of_week.reindex(days_of_week_order)

plt.figure(figsize=(10, 6))
average_orders_per_day_of_week.plot(kind='bar', color='skyblue')
plt.title('Average number of orders per day of the week')
plt.xlabel('Day of the week')
plt.ylabel('Average number of orders')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# ***Is there a day of the week that generated more revenue? ***

```python
revenue_per_day=df_merged.groupby('day_of_week')['revenue'].sum()
revenue_per_day=revenue_per_day.reindex(days_of_week_order)
```

```python
plt.figure(figsize=(10, 6))
revenue_per_day.plot(kind='bar', color='skyblue')
plt.title('Total revenue for a certain day')
plt.xlabel('Day of week')
plt.ylabel('Total revenue')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
